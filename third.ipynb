```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
plt.rcParams['font.sans-serif'] = ['SimHei']   #用来正常显示中文标签
plt.rcParams['axes.unicode_minus'] = False   #用来正常显示负号
```

```python
fig, axs = plt.subplots(2, 5, figsize=(10, 4), sharex=True, sharey=True)
fig.suptitle("样例1", size=20)
for i in range(2):
    for j in range(5):
        x = np.random.randint(-4, 5, 10)
        y = np.random.randint(-4, 5, 10)
        axs[i][j].scatter(x, y)
        axs[i][j].set_title("第%d行，第%d列" % (i + 1, j + 1))
        axs[i][j].set_xlim(-5, 5)
        axs[i][j].set_ylim(-5, 5)
        if i == 1:
            axs[i][j].set_xlabel("横坐标")
        if j == 0:
            axs[i][j].set_ylabel("纵坐标")
fig.tight_layout()
```

```python
plt.figure()
# 子图1
plt.subplot(2,2,1) 
plt.plot([1,2], 'r')
# 子图2
plt.subplot(2,2,2)
plt.plot([1,2], 'b')
#子图3
plt.subplot(224)  # 当三位数都小于10时，可以省略中间的逗号，这行命令等价于plt.subplot(2,2,4) 
plt.plot([1,2], 'g');
```

```python
N = 150
r = 2 * np.random.rand(N)
theta = 2 * np.pi * np.random.rand(N)  # 每一个点的偏移角度
area = 200 * r**2  # 每一个点的面积大小
colors = theta

plt.subplot(projection='polar')  # 极坐标的形式
plt.scatter(theta, r, c=colors, s=area, cmap='hsv', alpha=1);
```

```python
fig = plt.figure(figsize=(10, 4))
spec = fig.add_gridspec(nrows=2, ncols=5, width_ratios=[1,2,3,4,5], height_ratios=[1,3])
fig.suptitle('样例2', size=20)
for i in range(2):
    for j in range(5):
        ax = fig.add_subplot(spec[i, j])
        ax.scatter(np.random.randn(10), np.random.randn(10))
        ax.set_title('第%d行，第%d列'%(i+1,j+1))
        if i==1: ax.set_xlabel('横坐标')
        if j==0: ax.set_ylabel('纵坐标')
        
# 通过切片的方式达到子图合并的功能
ax = fig.add_subplot(spec[0, :3])
ax.scatter(np.random.randn(10), np.random.randn(10))

fig.tight_layout()
```

```python
fig, ax = plt.subplots(figsize=(4,3))
ax.axhline(0.5,0.2,0.8)  # 水平直线对应的y轴，直线x轴方向起点，直线x轴方向终点
ax.axvline(0.5,0.2,0.8)  # 竖直直线对应的x轴，直线y轴方向起点，直线y轴方向终点
ax.axline([0.4,0.3],[0.7,0.7]);  # 直线的起点坐标与终点坐标
ax.grid(True)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(10, 4))
for j in range(2):
    axs[j].plot([10, 100, 1000, 10000], [10**i for i in range(4)])
    if j == 0:
        axs[j].set_xscale('log')
    else:
        axs[j].set_yscale('log')
fig.tight_layout()
```
